```python
# 4. Examine the significance of the model
import statsmodels.api as sm
X = sm.add_constant(X)  # Adding a constant
model = sm.OLS(df['MEDV'], X).fit()
print(model.summary())
```

```python

# 5. Checking the below linear regression assumptions
# Mean of residuals
residuals = model.resid
print("Mean of residuals:", np.mean(residuals))

# No Heteroscedasticity
plt.scatter(model.fittedvalues, residuals)
plt.xlabel('Fitted values')
plt.ylabel('Residuals')
plt.title('Residuals vs Fitted values')
plt.show()

# Linearity of variables
sns.regplot(x=model.fittedvalues, y=np.sqrt(np.abs(residuals)), lowess=True)
plt.xlabel('Fitted values')
plt.ylabel('Sqrt(Abs(Residuals))')
plt.title('Scale-Location Plot')
plt.show()

# Normality of error terms
sns.histplot(residuals, kde=True)
plt.title('Residuals Distribution')
plt.show()
sm.qqplot(residuals, line ='45')
plt.title('QQ Plot')
plt.show()
```
